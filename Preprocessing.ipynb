# Preprocessing

## image Selection

```python
#1인당 60장씩 데이터로 사용할만한 이미지만 골라서 다른이름으로 저장

import shutil
import os 

#root_path = "C:/koreanface/middle400/middle400_1" #path는 역슬래쉬 말고 /로 나눈다 (C:\test => 이거 안됨)
destination_root_path = "kfaceData"

for i in range (1,13):
    root_path = "middle400/middle400_"+str(i)
    for person_name in (os.listdir(root_path)):
        copy = 0
        try:
            os.mkdir(os.path.join(destination_root_path, person_name))
            for S in [1,2,5]: #S001~S003 or S001,S002,S005(모자)
                for L in range(1,3): #L1~L2
                    for E in range(1,3): #E01~E02
                        for C in range(5,10): #C5~C9 
                            copy+=1
                            source_path = os.path.join(root_path, person_name, "S00"+str(S),"L"+str(L),"E0"+str(E),"C"+ str(C) + ".jpg") #원본파일 위치
                            if copy < 10:
                                destination_path = os.path.join(destination_root_path, person_name,str(person_name)+"_000"+str(copy)+".jpg") #복붙해서 넣어둘 위치
                            else:
                                destination_path = os.path.join(destination_root_path, person_name,str(person_name)+"_00"+str(copy)+".jpg")
                            shutil.copyfile(source_path, destination_path) #복붙 (copyfile함수는 path가 \가 아니라 /로 나누어져 있어야함)

        except Exception as err:
            print(err)
```

## face Detection

```python
import face_recognition
import argparse
import numpy as np
import os.path
import cv2 
```

```python
old_dir = 'kfaceData'
new_dir = 'faceData\kfaceData_detect'
```

```python
old_dir_list = os.listdir(old_dir)
len(old_dir_list)
```

```python
for root, dirs, files in os.walk(old_dir):
#     print(dirs)
    people_name = os.path.split(root)[1]
    print(people_name)
    new_people_dir = os.path.join(new_dir, people_name)
    try:
        os.mkdir(new_people_dir)
    except Exception as err:
        print(err)
        
    # face recognition and  save
    for i, file in enumerate(files):
        input_image_path = os.path.join(root, file)
        original_image = cv2.imread(input_image_path, 1)
        if original_image is not None:
            original_image= cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
            copy_image = original_image
            face_locations = face_recognition.face_locations(original_image)

            for face_location in face_locations:
                # more bigger box size than ordinary box
                top, right, bottom, left = face_location
                top -= 30
                right += 30
                bottom +=30
                left -=30
                
                crop_image = copy_image[top:bottom, left:right]
                resize_image = cv2.resize(crop_image, dsize=(220,220), interpolation = cv2.INTER_AREA)
                resize_image = cv2.cvtColor(resize_image, cv2.COLOR_RGB2BGR)
                save_path = f"{new_dir}/{people_name}/{file}"
                cv2.imwrite(save_path, resize_image)
```
